### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  
**4.1 Answer:**

Our overall framework follows the **Machine Learning Project Checklist** from *Hands-On Machine Learning* (HOML, Appendix A). The workflow proceeds in five sequential phases: (1) problem framing and EDA, (2) data cleaning and preprocessing, (3) feature engineering, (4) model selection and cross-validated evaluation, and (5) hyperparameter tuning and final model selection. Each phase fed directly into the next, and earlier decisions were revisited after observing model behavior in later phases.

**Validation Strategy:** Rather than a simple train/validation/test split, we used **K-Fold Cross-Validation** (5 splits, shuffled, `random_state=42`) throughout the model evaluation phases. This provides 5 independent estimates of generalization error from the training data alone, making the reported CV MAE far more reliable than a single hold-out validation score. A fixed 80/20 train-test split was created at the start of Milestone 2; the test set was held out entirely until the final model was selected — it was never used to make modeling decisions, only to produce the final reported test performance.

**Error Metric — MAE vs. RMSE:** We selected **Mean Absolute Error (MAE)** as our primary metric rather than RMSE for two reasons. First, MAE reports the average prediction error in the same units as the target (dollars), making it immediately meaningful to non-technical stakeholders. Second, the target distribution has a long right tail of genuinely high-value properties; RMSE, which squares errors before averaging, would be dominated by those extreme cases and give a distorted picture of typical performance. MAE is more robust to outliers and better reflects the average user experience on the listing platform. $R^2$ was tracked as a supplementary metric to characterize the overall proportion of variance explained.


### 4.2 Data Cleaning and Preprocessing

**4.2 Answer:**  
**4.2 Answer:**

The raw Zillow dataset required substantial cleaning before any modeling could begin. The primary issues were an excess of identifier columns, extreme rates of missing data, and a skewed target distribution.

**Dropping Uninformative Columns:** Any column whose name contained "id", "census", "fips", "desc", or "code" was dropped — these are row identifiers, geographic lookup codes, or text descriptions that cannot be meaningfully used without external reference tables. Latitude and longitude were also removed at this stage (a decision later reconsidered — see Section 4.4). Columns with fewer than 2 distinct values (excluding pool and garage counts) were also removed as near-constant features. This step eliminated roughly half of all columns while retaining all substantive physical and structural features.

**Handling Remaining Missing Values via Imputation:** Three strategies were applied based on the semantic meaning of missingness:
- **Zero-fill:** `poolcnt`, `garagecarcnt`, and `garagetotalsqft`. A null in these columns almost universally means the property has no pool or garage — not that the data was not recorded.
- **Median-fill:** `calculatedbathnbr`, `bathroomcnt`, `fullbathcnt`, `lotsizesquarefeet`, `calculatedfinishedsquarefeet`, and `yearbuilt`. The median is preferred over the mean for skewed distributions because it is not pulled by extreme values.
- **Mode-fill:** `numberofstories` and `unitcnt`, where the most common value represents the typical property.

After targeted imputation, any rows with remaining nulls were dropped. This preserved the vast majority of observations.

**What Did Not Work:** Retaining and encoding the high-cardinality categorical columns (`propertyzoningdesc`, `propertycountylandusecode`) added noise without improving model performance in preliminary experiments, so they were excluded. Dropping latitude and longitude, while reasonable at the time, was later identified as the most consequential data cleaning decision — location is a dominant predictor of property value and its absence contributes meaningfully to remaining prediction error.


### 4.3 Feature Engineering

Describe your feature engineering.

Include:
- transformations or new features
- why you created them
- which were useful or not useful
- what worked and what did not work

**4.3 Answer:**  
**4.3 Answer:**

Feature engineering was explored systematically in both milestones. Five new features were added to the baseline cleaned feature set, guided by the EDA findings that square footage exhibits a power-law (non-linear) relationship with assessed value and that ratio features can expose signals that raw counts cannot.

| Feature | Formula | Rationale |
|---------|---------|----------|
| `log_sqft` | log(1 + `calculatedfinishedsquarefeet`) | Compresses right-skew; linearizes the sqft-to-price power-law relationship |
| `log_lotsize` | log(1 + `lotsizesquarefeet`) | Same skew correction for lot size, which has a long right tail |
| `sqft_squared` | `calculatedfinishedsquarefeet`² | Captures possible convex price acceleration for very large homes |
| `property_age` | 2016 − `yearbuilt` | More interpretable depreciation signal than raw calendar year |
| `bath_to_bed_ratio` | `bathroomcnt` / `bedroomcnt` | Proxy for luxury/finish level beyond raw counts |

**What worked:** `log_sqft` and `property_age` ranked consistently in the top-10 features by both Random Forest importance and F-statistic, validating the engineering decisions. `bath_to_bed_ratio` provided modest gains for the ensemble models by capturing the luxury signal that neither count alone expresses.

**What did not work:** `sqft_squared` ranked low in importance and was dropped during feature selection. Tree-based models already learn non-linear splits internally, making a polynomial term redundant. Counterintuitively, the log transforms that benefit linear models have a more limited effect on ensemble methods — trees split on rank order, not scale, so the distribution shape matters less to them.


### 4.4 Model Selection

Describe how you selected your model.

Include:
- models you tried and why
- how you evaluated generalization
- hyperparameter tuning
- how you chose the final model

**4.4 Answer:**  
**4.4 Answer:**

**Models Considered:** Three regression algorithms were evaluated:

1. **Linear Regression** — an interpretable baseline. If the relationships in the data are approximately linear, a well-tuned linear model can match more complex methods while being easier to explain. It also served as a reference: if ensemble models did not clearly outperform it, the added complexity would not be justified.
2. **Random Forest Regressor** — an ensemble of independently grown decision trees. Handles non-linear interactions natively, is robust to outliers, and provides a built-in feature importance ranking used to guide feature selection.
3. **Gradient Boosting Regressor** — a sequential boosting ensemble where each successive tree fits the residuals of previous ones. Typically achieves lower bias than Random Forest on tabular regression tasks at the cost of more hyperparameters to tune.

**Evaluating Generalization:** All models were evaluated using 5-fold cross-validation on the training set with a fixed random seed. The test set was held out until the very end and used only once to confirm generalization. This strict separation prevents the optimistic bias that arises from using test performance to guide modeling decisions.

**Hyperparameter Tuning:**
- *Linear:* Replaced by **Ridge Regression** with `GridSearchCV` over regularization strengths (α from 0.01 to 1000). Ridge shrinks correlated coefficients without zeroing them out, addressing multicollinearity among bathroom-related features.
- *Random Forest:* `RandomizedSearchCV` over `n_estimators`, `max_depth`, `min_samples_leaf`, and `max_features`.
- *Gradient Boosting:* `RandomizedSearchCV` over `n_estimators`, `learning_rate`, `max_depth`, `subsample`, and `min_samples_leaf`. Lower learning rates (0.05–0.1) with more trees consistently outperformed aggressive rates.

**Final Model Selection:** The tuned **Gradient Boosting Regressor** was selected because it achieved the lowest CV MAE across all experimental stages, and its test MAE closely tracked the CV estimate — confirming genuine generalization. One key modeling decision reviewed in hindsight: latitude and longitude were dropped in preprocessing under the assumption that raw coordinates would be too granular without a neighborhood lookup. In reality, geographic coordinates are among the strongest predictors of property value in real-estate datasets, and their absence is the primary driver of the remaining prediction error. In a production scenario, these features would be retained.
